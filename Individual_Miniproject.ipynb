Data Cleaning

```python
# data cleaning: filling the missing data in 'Votes'
df.isna().sum().sort_values(ascending=False)
```

```python
# 'Votes' has 12 missing data. I chose to replace the missing data with the average number of votes.
df['Votes'].fillna(df['Votes'].mean(), inplace=True)
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
# data cleaning: converting 'Season' to integer
df.dtypes.sort_values(ascending=False)
```

```python
df['Season'] = pd.to_numeric(df['Season'], errors='coerce').astype('Int64')
df['Season'].info()
df.isna().sum().sort_values(ascending=False)
```

```python
df[df['Season'].isna()][['Season']]
```

```python
# Convert to string and extract digits
df['Season'] = df['Season'].astype(str).str.extract(r'(\d+)')

# Convert the extracted digits to numeric (nullable integer)
df['Season'] = pd.to_numeric(df['Season'], errors='coerce').astype('Int64')

# Replace the 24 missing values (NaN) with 2
df['Season'].fillna(2, inplace=True)

# Check that no missing values remain
print(df['Season'].isna().sum())

# Check the column type
print(df['Season'].dtype)


```

Reflection on Data Cleaning:

In 'Votes', only 12 values were missing, thus I decided to preserve the entire column by filling the missing data. I chose to fill them with the average value of 'Votes', in order to not bias the column to extreme values while assuming that the 12 missing values would not have been too different from the average.

In 'Season', I first converted the column to integers. Then I inspected the values that couldn't be converted. I used a regular expression to capture the numeric values in the problematic data, which is "2". I filled the missing data with "2".

data visualizaiton


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average Stars per director
director_ratings = df.groupby('Director')['Stars'].mean().sort_values(ascending=False).reset_index()

# Optional: only show directors with multiple episodes to avoid skew
episode_counts = df['Director'].value_counts()
directors_multiple_eps = episode_counts[episode_counts > 1].index
director_ratings = director_ratings[director_ratings['Director'].isin(directors_multiple_eps)]

# Plot with enhanced visual contrast
plt.figure(figsize=(14,8))
ax = sns.barplot(data=director_ratings, x='Stars', y='Director', palette='coolwarm_r', dodge=False)

# Add the rating value next to each bar
for p in ax.patches:
    width = p.get_width()  # the height of the bar (rating)
    ax.text(width + 0.05, p.get_y() + p.get_height() / 2, f'{width:.2f}',
            ha='left', va='center', fontsize=12, color='black')

# Customize plot
plt.xlabel('Average Rating (Stars)', fontsize=14)
plt.ylabel('Director', fontsize=14)
plt.title('Directors by Average Friends Episode Rating', fontsize=16)
plt.xlim(5, 10)

plt.show()
```

**Question 2 (5pt): Perform Clustering**

1.  **K-Means**
    *   Using the `Time_in_Art` and `Time_in_History` columns, apply the K-Means clustering algorithm with `random_state=42`. Set the number of clusters (`k`) to 3.
    *   Create a new column in your dataset called `Cluster` and assign the cluster label (e.g., 0, 1, or 2) to each visitor.

2.  **Visualize the Results:** Remake the scatter plot from question 1, but this time, color the points according to their assigned `Cluster`.

3. **Interpretation**: Reflect on how K-Means assigns clusters. Does the algorithm's result match your observations from the previous task?

Please add all your code to the next cell or create as many cells as you need.

```python
#K-Means

X = museum_visitors[['Time_in_Art', 'Time_in_History']]

kmeans = KMeans(n_clusters=3, random_state=42)
museum_visitors['Cluster'] = kmeans.fit_predict(X)

museum_visitors.head()
```

```python
# Visualize the results
plt.figure(figsize=(5, 5))
sns.scatterplot(data=museum_visitors, x='Time_in_Art', y='Time_in_History', hue = 'Cluster', alpha=0.7)
plt.title('Scatterplot of Time_in_Art vs Time_in_History')
```

3. **Interpretation**: What can you say about the persona's? Does their age relate to their cluster?

```python
# Profile the clusters
museum_visitors.groupby('Cluster')[['Age', 'Time_in_Art', 'Time_in_History']].mean()
```

```python
# Assign persona names

persona_map = {
    0: "Art Oriented",
    1: "Balanced Visitor",
    2: "History Oriented"
}

museum_visitors['Persona'] = museum_visitors['Cluster'].map(persona_map)

museum_visitors[['Cluster',  'Age', 'Persona']]
```
